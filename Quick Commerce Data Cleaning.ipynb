### 2) Rename Columns

```python
df = df.rename(columns= {'Delivery Time(Mins)': 'Delivery_Time(Mins)'})
df = df.rename(columns= {'Distance (km)': 'Distance_(km)'})
```

### 3) Correct Datatype

```python
def safe_fix_number(val):
    try:
        if pd.isna(val):
            return np.nan
        if isinstance(val, str):
            parts = val.split('.')
            if len(parts) > 2:
                # Remove second and further dots
                return float(parts[0] + '.' + ''.join(parts[1:]))
            return float(val)
        return float(val)
    except:
        return np.nan  # return NaN if conversion fails
```

```python
df['Restaurant_latitude'] = df['Restaurant_latitude'].apply(safe_fix_number)
df['Restaurant_longitude'] = df['Restaurant_longitude'].apply(safe_fix_number)
df['Delivery_location_latitude'] = df['Delivery_location_latitude'].apply(safe_fix_number)
df['Delivery_location_longitude'] = df['Delivery_location_longitude'].apply(safe_fix_number)
```

```python
df['Delivery_Time(Mins)'] = df['Delivery_Time(Mins)'].apply(safe_fix_number)
```

```python
df['Delivery_Time(Mins)'] = pd.to_numeric(df['Delivery_Time(Mins)'])
```

```python
latlong_cols  = ['Restaurant_latitude', 'Restaurant_longitude', 'Delivery_location_latitude','Delivery_location_longitude']
df[latlong_cols]= df[latlong_cols].apply(pd.to_numeric, errors = 'coerce')
```

```python
cat_cols = ['Traffic_Level', 'weather_description', 'Type_of_order', 'Type_of_vehicle']
df[cat_cols] = df[cat_cols].astype('category')
```

### 4) Handle Missing Values

```python
df.isnull().sum()
```

```python
# Fill categorical with mode
df['Traffic_Level'] = df['Traffic_Level'].fillna(df['Traffic_Level'].mode())
df['weather_description'] = df['weather_description'].fillna(df['weather_description'].mode())
```

```python
#Fill numeric with mean
df['temperature'] = df['temperature'].fillna(df['temperature'].mean())
df['humidity'] = df['humidity'].fillna(df['humidity'].mean())
df['precipitation'] = df['precipitation'].fillna(df['precipitation'].mean())
df['Delivery_Time(Mins)'] = df['Delivery_Time(Mins)'].fillna(df['Delivery_Time(Mins)'].median())
```

```python
# Fill distance using the Haversine formula( calculate distance using lattitudes and longitudes)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return R * 2 * np.arcsin(np.sqrt(a))

mask = df['Distance_(km)'].isnull()
df.loc[mask, 'Distance_(km)'] = df.loc[mask].apply(
    lambda row: haversine(
        row['Restaurant_latitude'], row['Restaurant_longitude'],
        row['Delivery_location_latitude'], row['Delivery_location_longitude']
    ),
    axis=1
)
mask = df['Distance_(km)'].isnull()
df.loc[mask, 'Distance_(km)'] = df.loc[mask].apply(
    lambda row: haversine(
        row['Restaurant_latitude'], row['Restaurant_longitude'],
        row['Delivery_location_latitude'], row['Delivery_location_longitude']
    ),
    axis=1
)
```
